```python
import pandas as pd
import numpy as np
```

```python
years = ['2019', '2020']
months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
data_df = pd.DataFrame()
for year in years:
    for month in months:
        if year == '2019':
            data_path = "Resources/JC Data/JC-" + year + month + "-citibike-tripdata.csv"
            aux_df = pd.read_csv(data_path)
            data_df = data_df.append(aux_df)
        elif (year == '2020') & (month != '12'):
            data_path = "Resources/JC Data/JC-" + year + month + "-citibike-tripdata.csv"
            aux_df = pd.read_csv(data_path)
            data_df = data_df.append(aux_df)
data_df
```

```python
data_df['gender'].replace(0, 'Unknown',inplace=True)
data_df['gender'].replace(1, 'Male',inplace=True)
data_df['gender'].replace(2, 'Female',inplace=True)
data_df.head()
```

```python
data_df['date'] = data_df['starttime'].str[:10]
data_df.sort_values('starttime')
data_df.head()
```

```python
data_df.to_csv(r'Clean Data/Citibike_Data.csv')
```

```python
start_data = data_df.copy()
start_data = start_data.drop(['tripduration', 'starttime', 'stoptime', 'end station name', 'end station longitude',
                         'bikeid', 'usertype', 'birth year', 'gender', 'date', 'end station id', 'end station latitude'], 1)

start_data = start_data.groupby(['start station id', 'start station name', 'start station latitude', 'start station longitude']).size().reset_index(name='start count')
start_data = start_data.sort_values('start count')
start_data = start_data.rename(columns={'start station id': 'station id', 'start station name': 'station name',
                                       'start station latitude': 'station latitude', 'start station longitude': 'station longitude'})
start_data
```

```python
end_data = data_df.copy()
end_data = end_data.drop(['tripduration', 'starttime', 'stoptime', 'start station name', 'start station longitude',
                         'bikeid', 'usertype', 'birth year', 'gender', 'date', 'start station id', 'start station latitude'], 1)

end_data = end_data.groupby(['end station id', 'end station name', 'end station latitude', 'end station longitude']).size().reset_index(name='end count')
end_data = end_data.sort_values('end count')
end_data = end_data.rename(columns={'end station id': 'station id', 'end station name': 'station name',
                                       'end station latitude': 'station latitude', 'end station longitude': 'station longitude'})
end_data
```

```python
stations_df = pd.merge(start_data, end_data, how='outer', on=['station name', 'station id', 'station longitude', 'station latitude'])
stations_df = stations_df.replace(np.nan, 0)
stations_df
```

```python
stations_df.to_csv(r'Clean Data/Stations_Data.csv')
```
